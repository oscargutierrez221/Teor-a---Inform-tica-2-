**¿Qué se verá durante la primera parte del curso?**

Durante la primera parte del curso se abordara la teoría y practica sobre el lenguaje de programación C++. Esto incluyendo variables, iteradores, bucles, condicionales, etc.

**C++ y Python**

C++ es un lenguaje de programación que se utiliza para crear software, videojuegos, sistemas operativos, etc. Python es un lenguaje de programación que se utiliza para crear software, videojuegos, sistemas operativos, etc.
A pesar de que son dos lenguajes de programación ambos son lenguajes de alto nivel y sus practicas son bastante trasversales, como por ejemplo: Los operadores matematicos. Más sin embargo, c++ te permite manejar la memoria de manera directa, lo que te permite crear programas mas eficientes. Algo que también ocurre en python, pero no es manipulable y por eso no nos dabamos cuenta de que se podía realizar.

**¿Qué se verá el la segunda parte del curso?**

Se vera el paradigma de programación orientada a objetos, que es un paradigma de programación que se utiliza para crear software, videojuegos, sistemas operativos que permite crear objetos con sus propiedades y metodos. Un paradigma que es bastante utilizado en la industria de la programación y mucho más en C++-
